### Aufgabe 1 : verallgemeinerte Fibonacci Zahlen

Schreiben Sie eine Funktion `AllgFib(F0=1,F1=1,a=1,b=1,N=3)`, welche die ersten `N` Folgeglieder der verallgemeinerten Fibonaccifolge

$$
    F_{n+1} = a F_n + b F_{n-1}
$$

mit den Anfangsbedingungen $F_0=$ `F0` und $F_1=$ `F1` berechnet und als Liste ausgibt. (Achtung : Die Definition in der Vorlesung war etwas anders)

Schreiben Sie eine Funktion `my_sqrt(x)`, welche mit Hilfe der verallgemeinerten Fibonacci Zahlen die Wurzel aus `x` bis auf Maschinengenauigkeit ausrechnet. Berechnen Sie $\sqrt{2}$ mit dieser Funktion. 

*Hinweis :* Iterieren Sie die Fibonacci Folge schrittweise bis sich das Verhaeltnis aufeinanderfolgender Zahlen nicht mehr aendert.

````python
def AllgFib(F0=1,F1=1,a=1,b=1,N=3): # Ich definiere eine Funktion, die mir die Fibonacci-Folge ausgeben soll
    r=[]                            # Postionsargumente: F0=1, F1=1 als Anfangsbedingungen, Schlüsselwortargumente: a=Faktor der ersten Zahl, b=Faktor der zweiten Zahl, N=Anzahl der angegebenen Folgenflieder nach F0 unf F1 
    for i in range(0,N):            # r ist die Liste, die durch die Folgende Schleife gefüllt wird, im Intervall zwischen 0 und dem vorher bestimmten N
        if i == 0:                  # für die erste Stelle i=0 wird die Anfangsbedingung F0 hinzugefügt
            r.append(F0)
        elif i == 1:                # für die zweite Stelle i=1 wird die Anfangsbedingung F1 hinzugefügt, auf diesen beiden Bedingungen baut dann der Rest der Folge auf:
            r.append(F1)
        else:                       # Alle weiteren Stellen werden nach Regel der Fibonacci_Folge hinzugefügt, wobei die letzte und vorletzte Stelle von r jeweil addiert werden
            r.append(a*r[-1]+b*r[-2])
        
    return r                        # die Liste r wird für die Funktion AllgFib ausgegeben
        
    
    
````

````python
# Ich definiere eine Funktion, die mir basierend auf der Fibonacci-Folge die Quadratwurzel jeder beliebigen Zahl x berechnen kann. 
# dazu gilt folgender Zusammenhang: (F_n+1/F_n)=a/2+-(sqrt(b+(a/2))) 
# Also wähle ich für a: a=2 und für b: b=x-1 (mit x als Zahl, aus der die  Wurzel gezogen wird)
# daraus folgt: (F_n+1/F_n)-1=sqrt(b+1)=sqrt(x)

def my_sqrt(x): # x als Zahl unter der Wurzel, also b+1
    n = 3
    messu = 10**-16 # Maschinengenauigkeit aus einem Vorlesungsskript entnommen  
    diff = (AllgFib(a=2,b=x-1,N=n)[-1]/AllgFib(a=2,b=x-1,N=n)[-2])-(AllgFib(a=2,b=x-1,N=n)[-2]/AllgFib(a=2,b=x-1,N=n)[-3])
    while abs(diff)>messu : #wenn die differenz der Quotienten aus den letzten beiden und den vorletzten beiden Fibonacci-Zahlen gleich der Maschinenungenauigkeit wird, so kann die Quadratwurzel nicht mehr genauer berechnet werden, die Mschinengenauigkeit ist erreicht
        n = n +1 # Der Vorgang soll so oft wiederholt werden, bis die Maschinenungenauigkeit erreicht ist. Dabei muss in der Schleife N=n solange erhöht werden, bis dieser Fall eintritt (schrittweises Iterieren, bis gilt: abs(diff)=messu) 
        diff = (AllgFib(a=2,b=x-1,N=n)[-1]/AllgFib(a=2,b=x-1,N=n)[-2])-(AllgFib(a=2,b=x-1,N=n)[-2]/AllgFib(a=2,b=x-1,N=n)[-3])
        print(n) #wenn n mit ausgegeben wird, kann man sehen über welche Zahlen die Quadratwurzel des berechnenten x berechnet wurde, bzw. die Länge des verwendeten Teils der Fibonacci-Folge
    
    wurz = (AllgFib(a=2,b=x-1,N=n)[-1]/AllgFib(a=2,b=x-1,N=n)[-2])-1 # Hier der Zusammenhang zwischen Fibonacci und der Quadratwiúrzel, um die Quadratwurzel zu erhalten
    return wurz
````

### Aufgabe 2 : Collatz-Problem

Wir betrachten die folgende Abbildung ueber den *positiven* natuerlichen Zahlen

$$
    f(n) = \left\lbrace \begin{array}{ll}
                            {}n/2 \qquad &,\textrm{falls }n\textrm{ gerade} \\
                            3n+1\qquad &,\textrm{falls }n\textrm{ ungerade}
                        \end{array}
           \right.
$$

sowie die natuerliche Zahl $\tau(n)$ welche angibt, wie of $f$ iteriert werden muss, bis man die Zahl 1 erreicht, d.h.

$$
    \left(f\circ\right)^{\tau(n)}(n) = 1
$$

Schreiben Sie eine Funktion `tau(n)`, welche diese Zahl berechnet. Berechnen Sie eine Liste `L` der Werte $\tau(n)$ fuer die ersten `2**20` positiven natuerlichen Zahlen und plotten diese mit

```python
    from pylab import *
    plot(L,'.') # nur Punkte setzen, ohne diese zu verbinden
    show()
```

<a href="https://en.wikipedia.org/wiki/Collatz_conjecture">Collatz-Problem @ Wikipedia</a><br>
Collatz-Problem @ xkcd : <img src="http://imgs.xkcd.com/comics/collatz_conjecture.png">

````python
def manfred(n): # manfred(n) ist die Funkton, die mir tau ausgibt, tau sei die Anzahl der Ausführungen der folgenden Schleife, bis n eins wird. 
    tau = 0     # wenn n=1, so wird die folgende Schleife nicht ausgeführt, dann gilt tau=0, da kein weiterer Schritt nötig ist, um 1 zu erreichen.
    while n!=1: # die Schleife soll bei n=1 stoppen, da sich die Werte 4;2 und 1 danach unendlich wiederholen würden.
        if n%2==0: # Wenn n gerade, also durch 2 teilbar ist, soll durch 2 geteilt werden.
            n=n/2
            tau = tau+1 # jedes mal, wenn durch 2 geteilt werden kann, soll 1 zu tau addiert werden, da tau die Schritte, bis n=1 erreicht zählt.
        else:
            n=n*3+1 # ist n ungerade, soll n*3+1 ausgeführt werden
            tau = tau+1 # auch hier muss ein weiterer Schritt als tau+1 gezählt werden. 
    return tau #tau, also die Anzahl der Schritte, bis n=1 erreicht wird, wird ausgegeben.
````

````python
print (manfred(9))
````

````python
def herbert(ö): # herbert(ö) definiert die Liste t, die die Funktion manfred für das Intervall von 1 bis ö ausgibt.  
    t=[manfred(n) for n in range(1,ö+1)]
    return t
````
