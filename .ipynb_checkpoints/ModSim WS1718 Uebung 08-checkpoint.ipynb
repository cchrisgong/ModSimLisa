````python
alpha = np.linspace(0,2*pi,200)       # Vektoren von 0 bis 2pi mit 200 Werten
beta = np.linspace(0,2*pi,200)
Alpha, Beta = np.meshgrid(alpha,beta) # erstellt 2d Vektoren Alpha und Beta 


X = sin(Alpha) * cos(Beta)            # parameterisierung der Kugel
Y = sin(Alpha) * sin(Beta)
Z = cos(Alpha)

fig = figure('3D function wireframe plot')    
ax = fig.add_subplot(111,projection='3d')
wf = ax.plot_wireframe(X,Y,Z,rstride=5,cstride=5)   # Befehle zum Plotten der Kugel

ax.set_xlabel('$x$',fontsize=10)
ax.set_ylabel('$y$',fontsize=10)
ax.set_zlabel('$z$',fontsize=10)
ax.set_title('$Kugel$')                             
show()
````

````python
alpha = np.linspace(0,2*pi,200)                         # Vektoren von 0 bis 2pi mit 200 Werten 
beta = np.linspace(0,2*pi,200)

R = 2                                                   # Radius Mitte Torus bis Mitte Schlauch 
r = 1                                                   # Radius Mitte Schlauch bis Rand Schlauch
Alpha, Beta = np.meshgrid(alpha,beta)                       # Parameterisierung des Torus 
X = (R + r*cos(Alpha))*cos(Beta) 
Y = (R + r*cos(Alpha))*sin(Beta) 
Z = r*sin(Alpha)

fig=figure()
ax3d=fig.add_subplot(111,projection='3d',aspect=1)      # Plotten des Torus
ax3d.plot_wireframe(X,Y,Z,rstride=10,cstride=10)

ax3d.set_xlim(-(R+r+0.2),(R+r+0.2))
ax3d.set_ylim(-(R+r+0.2),(R+r+0.2))
ax3d.set_zlim(-(R+r+0.2),(R+r+0.2))

show()
````

````python

rad = np.linspace(-1,1,200)                         # Vektor von -1 bis 1 mit 200 Werten  
wink = np.linspace(0,2*pi,200)                      # Vektor von 0 bis 2pi mit 200 Werten

Rad, Wink = np.meshgrid(rad,wink)
X = cos(Wink) * (1 + (Rad/2) * cos(Wink/2))         # Parameterisierung des Möbiusbandes
Y = sin(Wink) * (1 + (Rad/2) * cos(Wink/2))
Z = (Rad/2) * sin(Wink/2)

fig=figure()
ax3d=fig.add_subplot(111,projection='3d',aspect=1)  # Plotten des Möbiusbandes
ax3d.plot_wireframe(X,Y,Z,rstride=10,cstride=10)

ax3d.set_xlim(-1,1)
ax3d.set_ylim(-1,1)
ax3d.set_zlim(-1,1)

show()
````

### Aufgabe 2 : Newton Fraktale

- Schreiben Sie eine Funktion `Newton_Steps(x0,f,fprime,dmin=1e-15,dmax=1e3,Nmax=100)` welche das Newton Verfahren mit der Funktion `f(x)` und deren Ableitung `fprime(x)` beginnend bei $x_0$ solange ausfuehrt bis $\left|f(x_n)\right|\le d_{min}$, $\left|f(x_n)\right|\ge d_{max}$ oder $n=N_{max}$ und den Wert $x_n$ der Nullstellen-Naeherung sowie die Zahl $n$ der notwendigen Iterationsschritte ausgibt.

- Bestimmen Sie fuer alle *komplexen* Anfangswerte $z =x+iy $ auf einem geeigneten Gitter die Approximation der von dort gefundenen Nullstellen der Funktion $f(z)=z^3-1$ und die Zahl der benoetigten Iterationsschritte. Berechnen Sie aus diesen Werten geeignete `[r,g,b]` Farben oder eine skalare Funktion und plotten Sie diese mit `imshow` ueber dem Gitter.

Hinweis : Wenn Sie die Funktionen `Newton_Steps`, `f` und `fprime` sorgfaeltig programmieren, koennen diese direkt auf ein ganzes `numpy` Feld von Anfangswerten gleichzeitig angewendet werden ohne Python Schleifen zu verwenden.

````python
from pylab import *

def Newton_Steps(x0,f,fprime,dmin=1e-15,dmax=1e3,Nmax=100):
    n = 0
    nullx = x0                                                   # Anfangspunkt
    nully = f(nullx)                                             
    while abs(nully) > dmin and abs(nully) < dmax and n <= Nmax: # Abbruchbedingungen für Newtonverfahren 
        nullx = nullx - (f(nullx)/fprime(nullx))                 # Bildungsvorschrift 
        nully = f(nullx)                                         
                                                                 
        n = n + 1
        
    return (nullx,n) 
````

````python
def f(z):                                        # Definieren der Funktion f(z)               
    return z**3 - 1.0

                           

N=200                                            
eps=1e-5                                         # Genauigkeit 
h=1e-5                                          
maxIt=20                                         # Höchste Anzahl der Durchläufe
M = np.zeros((N,N,3))                            # Matrix für Fraktal
a = linspace(-1,1,N)                          # reelle Achse
b = linspace(-1,1,N)                          # imaginäre Achse
for i in range(0,N):                             
    for j in range(0,N):                         
        n = 0                                    # Zahl der Durchläufe
        z = complex(a[i],b[j])                   # Komplexe Zahl, abhängig von reeller und imaginärer Achse
        pruefr = f(z).real                       # Realteil f(z)
        pruefi = f(z).imag                       # imaginärteil f(z)
        while abs(pruefr) > eps and abs(pruefi) > eps and n <= 20:  # Abbruchbedingungen Newtonverfahren 
                                                        
                        
            dz = (f(z + complex(h,h)) - f(z)) / complex(h,h)        # nummerische Ableitung von f(z) 
            z0 = z - f(z)/dz                                        # Bildungsvorschrift für z0
            n = n + 1                                               
            z = z0 
            pruefr = f(z).real                                      # neu Berechnung der Abbruchbedinguhngen 
            pruefi = f(z).imag
       
        n1r = z.real - 1                                            # Vergleich des Startpunktes mit den drei möglichen Nullstellen
        n1i = z.imag - 0                                            
        n2r = z.real - cos(2*pi/3)
        n2i = z.imag - sin(2*pi/3)
        n3r = z.real - cos(4*pi/3)
        n3i = z.imag - sin(4*pi/3)
        
        if abs(n1r) < 1e-3 and  abs(n1i) < 1e-3:                    # rot für erste Nullstelle
            M[j,i,0] = (maxIt+1)/(1/(n + 1))                       
            M[j,i,1] = 0                                           
            M[j,i,2] = 0                                           
        elif abs(n2r) < 1e-3 and  abs(n2i) < 1e-3:                  # grün für zweite Nullstelle
            M[j,i,0] = 0
            M[j,i,1] = (maxIt+1)/(1/(n + 1))
            M[j,i,2] = 0
        elif abs(n3r) < 1e-3 and  abs(n3i) < 1e-3:                  # blau für dritte Nullstelle
            M[j,i,0] = 0
            M[j,i,1] = 0
            M[j,i,2] = (maxIt+1)/(1/(n + 1))
        else:                                                       # schwarz falls keine Konvergenz
            M[j,i,0] = 0
            M[j,i,1] = 0
            M[j,i,2] = 0

plt.imshow(M,interpolation='none',origin='lower',extent=[-1,1,-1,1]) # Plotten des Fraktals 
show()
````
